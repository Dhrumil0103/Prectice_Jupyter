```python
# Plot 1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(2,2,1) # 2 Row, 2 Column and first plot
plt.plot(x,y, ls = "-.")

# Plot 2
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(2,2,2) # 2 Row, 2 Column and second plot
plt.plot(x,y,ls = "-.")

plt.show()
```

```python
# P 1
a = np.array([0,1,2,3])
b = np.array([3,8,1,10])

plt.subplot(2,1,1)
plt.plot(a,b, "o", ls = "-.")

# P 2
a = np.array([0,1,2,3])
b = np.array([10,20,30,40])

plt.subplot(2,1,2)
plt.plot(a,b, "o", ls = "--")

plt.show()
```
