```python
df = np.array([3,4,8,1])
gf = np.array([7,1,6,9])

plt.plot(df, gf, marker = "o")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
jg = np.array([5,9])
gd = np.array([2,6])

plt.plot(jg, gd)
plt.show()
```

```python
qw = np.array([5,3])
er = np.array([2,4])

plt.plot(qw, er, "D", c = "purple", ms = 11)
plt.show()
```

```python
gd = np.array([3,8])
df = np.array([1,9])

plt.plot(gd, df, "dr")
plt.show()
```

```python
gd = np.array([3,8])
df = np.array([1,9])

plt.plot(gd, df, "x-b")
plt.show()
```

```python
gd = np.array([3,8])
df = np.array([1,9])

plt.plot(gd, df, "H:g", ms = 10)
plt.show()
```

```python
gd = np.array([3,8])
df = np.array([1,9])

plt.plot(gd, df, "v--k")
plt.show()
```

```python
gd = np.array([3,8])
df = np.array([1,9])

plt.plot(gd, df, "^-.y")
plt.show()
```

```python
ncr = np.array([3,8,1,12])

plt.plot(ncr, marker = "p", ms = 17)
plt.show()
```

```python
ncr = np.array([3,8,1,12])

plt.plot(ncr, marker = "p", ms = 17, mec = "r") # Using mec - Markeredgecolor
plt.show()
```

```python
ncr = np.array([3,8,1,12])

plt.plot(ncr, marker = "p", ls = "-.",  c = "k", ms = 17, mfc = "r") # Using mfc - Markerfacecolor
plt.show()
```

```python
mf = np.array([3,8,7,15])

plt.plot(mf, "d", c = "purple", ls = "-.", ms = 13, mec = "r", mfc = "g")
plt.show()
```

```python
plt.plot(mf, "d", c = "purple", ls = "-.", ms = 13, mec = "#00cc00", mfc = "#cc0066")
plt.show()
```

```python
plt.plot(mf, "d", c = "purple", ls = "-.", ms = 13, mec = "greenyellow", mfc = "chocolate")
plt.show()
```

```python
plt.plot(mf, "d", c = "purple", ls = "-.", lw = "7.5", ms = 13, mec = "r", mfc = "g")
plt.show()
```

```python
ty = np.array([3,6,4,9])
yt = np.array([1,4,9,3])

plt.plot(ty, "-.g")
plt.plot(yt, "--r")
plt.show()
```

```python
x1 = np.array([0,1,2,3])
y1 = np.array([3,8,4,10])
x2 = np.array([0,1,2,3])
y2 = np.array([1,4,3,7])

plt.plot(x1, y1, x2, y2, marker = "s", lw = "3.5")
plt.show()
```

```python
plt.plot(x1, y1, "--D", c = "rebeccapurple", lw = 4.5, ms = "10")
plt.plot(x2, y2, "-.d", c = "g", lw = 2.5, ms = "7")
plt.show()
```
