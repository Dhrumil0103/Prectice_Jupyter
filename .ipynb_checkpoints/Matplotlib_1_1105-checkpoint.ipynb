```python
er = np.array([3,4,8,1])
we = np.array([7,1,6,9])

plt.plot(er,we)
plt.show()
```

```python
bn = np.array([5,9])
yu = np.array([2,6])

plt.plot(bn,yu)
plt.show()
```

```python
bn1 = np.array([5,3])
yu1 = np.array([2,4])

plt.plot(bn1, yu1, "o")
plt.show()
```

```python
hj = np.array([5,8,1,5])

plt.plot(hj)
plt.show()
```

```python
b1 = np.array([5,3])
n1 = np.array([2,4])

plt.plot(b1,n1,"$")
plt.show()
```

```python
bv = np.array([5,3])
cv = np.array([2,4])

plt.plot(bv, cv, marker = "*")
plt.show()
```

```python
nm = np.array([4,6,1,5])

plt.plot(nm, linestyle = "--", marker = "*")
plt.show()
```

```python
mn = np.array([4,6,1,5])

plt.plot(mn, ls = "-.", marker = "*", c = "red")
plt.show()
```

```python
fg = np.array([6,4,8])
nz = np.array([4,6,1,5])

plt.plot(fg, ls = "-.", c = "green")
plt.plot(nz, ls = "--", marker = "*", c = "red")
plt.show()
```

```python
gf = np.array([6,4,8])
mb = np.array([4,6,1,5])

plt.plot(gf, ls = "-.", c = "purple", lw = "5.5")
plt.plot(mb, ls = "--", marker = "*", c = "red", linewidth = "10.5")
plt.show()
```

```python
df = np.array([6,4,8])
jh = np.array([4,6,1,5])

plt.plot(df, ls = "-.", c = "black", lw = "7.5")
plt.plot(jh, ls = "--", marker = "*", linewidth = "20.5", c = "red")
plt.show()
```

```python
js = np.array([6,4,8])
sx = np.array([4,6,1,5])

plt.plot(js, ls = "-.", c = "blue", lw = "3.5")
plt.plot(sx, linestyle = "--", c = "darkorange", linewidth = "10.5", marker = "*")

plt.xlabel("Age")
plt.ylabel("Value")
plt.show()
```

```python
ad = np.array([6,4,8])
ds = np.array([4,6,1,5])

plt.plot(ad,ls = "-.", c = "gray", lw = "3.7")
plt.plot(ds, ls = "--", marker = "*", c = "darkgreen", lw = "10.5")

plt.xlabel("Age")
plt.ylabel("Value")

plt.grid()

plt.show()
```

```python
er = np.array([6,4,8])
rt = np.array([4,6,7,1])

plt.plot(er,"o", c = "green")
plt.plot(rt, "*", c = "red")
plt.grid()
plt.show()
```
