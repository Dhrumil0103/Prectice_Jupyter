## Matplotib Labels and Title

### Create Label for a Plot

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y, marker = "h")

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

### Create a Title for a Plot

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x, y, marker = "h")

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

### Set Font Properties for Title and Labels

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

f1 = {'family': 'serif','weight': 'bold','color': 'red','size': 20}
f2 = {'family': 'serif','color': 'darkred','size': 15}

plt.title("Sports Watch Data", fontdict = f1)
plt.xlabel("Average Pulse", fontdict = f2)
plt.ylabel("Calorie Burnage", fontdict = f2)

plt.plot(x,y)
plt.show()
```

### Position the Title

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x, y, marker = "h")

plt.title("Sports Watch Data", loc = "left")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```
